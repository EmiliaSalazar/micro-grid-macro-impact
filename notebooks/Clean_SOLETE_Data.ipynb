```python
def clean_SOLETE_data(file_path, resolution, delim=',', save_as_new_file = False):
    """
    Reduce time res
    """

    #Read csv and create df
    try:
        df = pd.read_csv(file_path,delimiter=delim)
    except Exception as e:
        print(f"Error reading file: {e}")
        return

    df = df.rename(columns = {'TEMPERATUREdegC | solete-temperature | 4053734': 'Temperature [ยบC]',
       'HUMIDITY | solete-humidity | 4053735' : 'Humidity',
       'WIND_SPEEDm1s | solete-windspeed | 4053736':'Wind Speed [m/s]',
       'WIND_DIRdeg | solete-winddir | 4053737' : 'Wind Direction [ยบ]',
       'GHIkW/m2 | solete-ghi | 4053738': 'GHI [kW/m2]',
       'POA IrrkW/m2 | solete-poairr | 4053739': 'POA Irr [kW/m2]',
       'P_GaiakW | solete-Pgaia | 4053740': 'Power Gaia [kW]',
       'P_SolarkW | solete-Psolar | 4053741' : 'Power Solar [kW]',
       'Pressurembar | solete-pressure | 4053742': 'Pressure [mbar]',
       'Azimuthdeg | solete-azimuth | 4053743': 'Azimuth [ยบ]',
       'Elevationdeg | solete-elevation | 4053744': 'Elevation [ยบ]'})
    
    df["ts"] = pd.to_datetime(df['ts'])
    df = df.set_index("ts") # set time as index
    df.reset_index("ts")
    
    df = df.resample(f'{resolution}min',).mean() # resize resolution
    
    if save_as_new_file == True:
        df.to_csv(f'{file_path[:-4]}_{resolution}min.csv', index=True) # save new file
    return df 

```

```python
df_SOLETE_15min = clean_SOLETE_data(SOLETE_data_1sec, 15, save_as_new_file = True)
```
