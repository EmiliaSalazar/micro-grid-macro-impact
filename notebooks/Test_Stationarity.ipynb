```python
from statsmodels.tsa.stattools import adfuller  
import pandas as pd
from pathlib import Path
```

```python
# Find repo root even if notebook is inside /notebooks
ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
#if ROOT.name == "micro-grid-macro-impact":
   # ROOT = ROOT.parent
file = str(ROOT / "data/combined_data_weather_and_power_30min.csv")
df = pd.read_csv(file)
```

```python
columns = ["Gaia_WT Power","Aircon_WT Power","B715","B117","B319","B330_1","B330_2","B330_3","B716"]
```

```python
for col in columns:  
    result = adfuller(df[col].dropna())
    pvalue = result[1]  
    if pvalue < 0.05:  print(f' {col} is stationary')  
    else:  print(f'{col} is not stationary')
```
