```python
import csv
import pandas as pd
from pathlib import Path
```

```python
def clean_weather_data(file_path,resolution,delim=",",save_as_new_file=False):
    """
    Reduce time resolution 
    ### NOTE: most of the variables were only added on the 15.06.2025
    """
    # Read the CSV file
    try:
        df = pd.read_csv(file_path,delimiter=delim)
    except Exception as e:
        print(f"Error reading file: {e}")
        return
    
    df["ts"] = pd.to_datetime(df['ts'],format="mixed")
    df = df.set_index("ts") # set time as index

    df = df.drop(columns=["_Meteo330M_MT_TEMPERATURE_temp_1 | syslab-06/meteo330/TEMPERATURE_temp_1 | 1205305",
                          "_Meteo330M_MT_HUMIDITY_hum_1 | syslab-06/meteo330/HUMIDITY_hum_1 | 1205306",
                          "_Meteo330M_MT_PRESSURE_pres_1 | syslab-06/meteo330/PRESSURE_pres_1 | 1205307",
                          "_Meteo330M_MT_WINDDIR_wdir_1 | syslab-06/meteo330/WINDDIR_wdir_1 | 1205308",
                          "_Meteo330M_MT_WINDSPEED_wspd_1 | syslab-06/meteo330/WINDSPEED_wspd_1 | 1205309",
                          "_Meteo330M_MT_INSOLATION_irrad_15w | syslab-06/meteo330/INSOLATION_irrad_15w | 1205310",
                          "_Meteo330M_MT_INSOLATION_irrad_0 | syslab-06/meteo330/INSOLATION_irrad_0 | 1205311",
                          "_Meteo330M_MT_INSOLATION_irrad_15e | syslab-06/meteo330/INSOLATION_irrad_15e | 1205312"])
    ## REMOVE all data for B330 and instead look at B715

    df = df.drop(columns=["_Meteo715M_MT_TEMPERATURE_temp1 | syslab-13/meteo1/TEMPERATURE_temp1 | 804140",
                          "_Meteo715M_MT_HUMIDITY_hum1 | syslab-13/meteo1/HUMIDITY_hum1 | 804141"])
    ## REMOVE temperature, and humidity??? I assume so

    df = df.resample(f'{resolution}min',).mean() # resize resolution
    if save_as_new_file == True:
        df.to_csv(f'{file_path[:-4]}_{resolution}min.csv', index=True) # save new file
    return df 
```
