```python
import csv
import pandas as pd
from pathlib import Path
```

```python
# Find repo root even if notebook is inside /notebooks
ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
#if ROOT.name == "micro-grid-macro-impact":
   # ROOT = ROOT.parent
SOLETE_prices = str(ROOT / 'data/Elspotprices_SOLETE.csv')
```

```python
def clean_old_format_day_ahead_data(file_path,delim=";",onlyDK2=True,timezone = "DK",save_as_new_file = False):
    """
    Reduce unnecessary columns (anything but DK2,UTC time zone)
    reverse the data frame, since it goes backwards in comparison to Syslab
    """
    if onlyDK2 == True: 
        region = ["DK2"]
    else: region = ["DK1","DK2"]
        
    # Read the CSV file
    try:
        df = pd.read_csv(file_path,delimiter=delim,decimal=",")
    except Exception as e:
        print(f"Error reading file: {e}")
        return
    

    df_DK = df[df["PriceArea"].isin(region)]
    df_DK = df_DK.drop(columns=["PriceArea","SpotPriceEUR"])

    if timezone == "DK":
        df_timecorrected = df_DK.drop(columns=["HourUTC"])
    else: df_timecorrected = df_DK.drop(columns=["HourDK"])

    df_timecorrected = df_timecorrected.rename(columns={"HourDK":"ts"},)
    df_reversed = df_timecorrected.iloc[::-1].reset_index(drop=True) # reverse time so it moves forward

    if save_as_new_file == True: 
        df_reversed.to_csv(f'{file_path[:-4]}_cleaned.csv', index=False, sep =",")

    return df_reversed
```

### **Clean SOLETE-relevant Day-Ahead Market Price Data**

```python
clean_old_format_day_ahead_data(SOLETE_prices, save_as_new_file = True)
```
