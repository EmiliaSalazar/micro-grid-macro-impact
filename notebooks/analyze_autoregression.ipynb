```python
from statsmodels.graphics.tsaplots import plot_acf
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
#if ROOT.name == "micro-grid-macro-impact":
   # ROOT = ROOT.parent
windspeed_file = str(ROOT / "data/DMI_ws.csv")
windspeed_df = pd.read_csv(windspeed_file)
irr_file = str(ROOT / "data/DMI_radiation.csv")
irr_df = pd.read_csv(irr_file)
PV_df = pd.read_csv(str(ROOT / "data/SyslabPV_15min_collective_PV.csv"))
Wind_df = pd.read_csv(str(ROOT / "data/SyslabWind_15min_nozeros.csv"))
Wind_df2 = pd.read_csv(str(ROOT / "data/complete_dataframe_30min.csv"))
```

```python
fig  = plot_acf(windspeed_df["Windspeed"],lags=200, title = "Autocorrelation of Windspeed") # no peak at 168 which would indicate weekly seasonality! but peaks only at 24*n, n being an integer 
# DAILY SEASONALITY
```

```python
fig  = plot_acf(irr_df["mean radiation"],lags=200,title= "Autocorrelation of Irradiance") # no peak at 168 which would indicate weekly seasonality! but peaks only at 24*n, n being an integer 
# DAILY SEASONALITY
```
